# Functions

```python
with open('./configs/labelconcat_config.yaml', 'r') as f:
        settings = yaml.load(f, Loader=yaml.FullLoader)
        settings['experiment']['name'] = 'real_nested/baselines'
xval_plotter = XvalLabelPlotter(settings)

with open('../data/experiment_keys/permutation_maps/vector_binary.yaml', 'rb') as fp:
    label_map = yaml.load(fp, Loader=yaml.FullLoader)
```

```python
def get_pngaddresses(addresses, df):
    ids = list(df.index)
    adds = []
    for iid in ids:
        adds = adds + ['./colour timelines/' + t for t in addresses if iid in t and '_t2.' in t]
    return adds
    

def get_highlow(experiment, xval_plotter):
    summary_df = xval_plotter.get_summary_df(experiment)

    summary_df['vector_labels'] = summary_df['permutation'].apply(lambda x: label_map['map'][x])
    summary_df['pos_proba'] = summary_df['probas'].apply(lambda x: x[1])
    
    high = summary_df[summary_df['vector_labels'] == '111']
    high_bad = high[high['pos_proba'] < 0.2]
    high_good = high[high['pos_proba'] > 0.8]

    low = summary_df[summary_df['vector_labels'] == '000']
    low_bad = low[low['pos_proba'] > 0.8]
    low_good = low[low['pos_proba'] < 0.2]

    print(
        'There are {} high performers that are predicted as low performers, and {} low performers that are predicted as high performers'
    .format(len(high_bad), len(low_bad)))
    
    timeline_folder = '../notebooks/5 - training/colour timelines/'
    timelines = os.listdir(timeline_folder)
    timelines = [t for t in timelines if t.endswith('.png')]
    
    hb_addresses = get_pngaddresses(timelines, high_bad)
    hg_addresses = get_pngaddresses(timelines, high_good)

    lb_addresses = get_pngaddresses(timelines, low_bad)
    lg_addresses = get_pngaddresses(timelines, low_good)
    
    # high ipyplot
    hb_labels = ['low (wrong)' for _ in hb_addresses]
    hg_labels = ['high (good)' for _ in hg_addresses]

    high_addresses = hb_addresses + hg_addresses 
    high_labels = hb_labels + hg_labels
    
    
    # low ipyplot
    lb_labels = ['high (wrong)' for _ in lb_addresses]
    lg_labels = ['low (good)' for _ in lg_addresses]
    low_addresses = lb_addresses + lg_addresses
    low_labels = lb_labels + lg_labels
    
    # high predicted low and low predicted low
    pred_low_labels = ['high (pred low)' for _ in hb_addresses] + ['low (pred low)' for _ in lg_addresses]
    pred_low_addresses = [adr for adr in hb_addresses] + [adr for adr in lg_addresses]
    
    # high predicted high and low predicted high
    pred_high_labels = ['high (pred high)' for _ in hg_addresses] + ['low (pred high)' for _ in lb_addresses]
    pred_high_addresses = [adr for adr in hg_addresses] + [adr for adr in lb_addresses]

    
    plot_data = {
        'high': {
            'addresses': high_addresses,
            'labels': high_labels
        },
        'low': {
            'addresses': low_addresses,
            'labels': low_labels
        },
        'pred_high': {
            'addresses': pred_high_addresses,
            'labels': pred_high_labels
        },
        'pred_low': {
            'addresses': pred_low_addresses,
            'labels': pred_low_labels
        }
    }
    
    return plot_data

    
```
