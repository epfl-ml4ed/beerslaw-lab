```python
# renaming
keras_path = '../../experiments/nested/lstm/language_colourbreak/binconcepts/lstm/raw_full/2022_01_23_0/0/logger/ctGRU_nlayers1_ncells[32]_drop02_optimadam_lossauc_bs16_ep50lstm/f9_model_checkpoint/'
checkpoint_paths = []
for (dirpath, dirnames, filenames) in os.walk(keras_path):
    files = [os.path.join(dirpath, file) for file in filenames]
    checkpoint_paths.extend(files)
```

```python
for path in checkpoint_paths:
    new_path = path.replace('variables/variables.', 'variables/variables/cp.ckpt.')
    new_folder = '/'.join(new_path.split('/')[:-1])
    os.makedirs(new_folder, exist_ok=True)
    os.rename(path, new_path)
```
