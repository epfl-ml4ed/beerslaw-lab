```python
import os
import re
import yaml
import pickle
import pandas as pd
import ipyplot

from visualisers.checkpoint_plotter import CheckpointPlotter
```

# Simple states 

```python
with open('./configs/plotter_config.yaml', 'r') as f:
    settings = yaml.load(f, Loader=yaml.FullLoader)
    settings['nocache'] = False
settings['experiment']['name'] = '/prior/yl_colourbreak/'


map_path = '../data/experiment_keys/permutation_maps/vector_binary.yaml'
with open(map_path) as fp:
    map_labels = yaml.load(fp, Loader=yaml.FullLoader)
```

```python
def explore_predsofperms(permutation:str):
    # Create plotter object
    plotter = CheckpointPlotter(settings)
    best_models = plotter.get_bestmodels()

    # Mapping from index to permutation, vector label and learner id and reverse key
    idx_permutations = {}
    lid_permutations = {}
    best_models['id_dictionary']['sequences']
    for idx in best_models['id_dictionary']['sequences']:
        with open(best_models['id_dictionary']['sequences'][idx]['path'], 'rb') as fp:
            sim = pickle.load(fp)
            idx_permutations[idx] = {
                'permutation': sim['permutation'],
                'vector_labels': map_labels['map'][sim['permutation']],
                'learner_id': best_models['id_dictionary']['sequences'][idx]['learner_id']
            }
            lid_permutations[best_models['id_dictionary']['sequences'][idx]['learner_id']] = {
                'idx': idx
            }

    # Predictions and probabilities for each of the validation folds
    y_predict = []
    y_probas = []
    val_indices = []
    vector_labels = []
    permutations = []
    ids = []
    for fold in best_models['folds']:
        val = list(best_models['folds'][fold]['val_indices'])
        ids = ids + [idx_permutations[idx]['learner_id'] for idx in val]
        val_indices = val_indices + list(best_models['folds'][fold]['val_indices'])
        vector_labels = vector_labels + [idx_permutations[idx]['vector_labels'] for idx in val]
        permutations = permutations + [idx_permutations[idx]['permutation'] for idx in val]

        y_predict = y_predict + list(best_models['folds'][fold]['y_predict'])
        y_probas = y_probas + [proba[1] for proba in best_models['folds'][fold]['y_proba']]

    cd ../notebooks/5 - training/
    # Dataframe
    predictions = pd.DataFrame([val_indices, vector_labels, permutations, y_predict, y_probas, ids]).transpose()
    predictions.columns = ['val_indices', 'vector_labels', 'permutations', 'y_predict', 'y_proba', 'learner_id']

    timelines = predictions[predictions['vector_labels'] == permutation]
    
    
    
    # Selecting the correct addresses and timelines
    timelines_path = './colour timelines/'
    addresses = os.listdir(timelines_path)
    addresses = [ad for ad in addresses if 'png' in ad]
    local_addresses = [(timelines_path + addr) for addr in addresses]
    lids = list(timelines['lids'])
    adds = []
    for lid in lids:
        adds = adds + [lad for lad in local_addresses if lid in lad and '_t2.png' in lad]
    lid_regex = re.compile('p2013_l([A-z0-9]+)_t2.png')
    lids = {lid_regex.findall(adx)[0]:adx for adx in adds}
    
    # Id to png address
    timelines['adress'] = timelines['lids'].apply(lambda x: lids[x])

    # plot
    ipyplot.plot_class_tabs(timelines['adress'], timelines['y_predict'], max_imgs_per_tab=10, img_width=700)

    
    
    
    
```
