```python
import os
import yaml
import dill
import pickle
import json

import pandas as pd
import numpy as np

from matplotlib import pyplot as plt
```

```python
cd ../../../src
```

```python
# demographics
with open('../data/beerslaw/post_test/rankings.pkl', 'rb') as fp:
    rankings = pickle.load(fp)
    rankings = rankings.set_index('username')
    
# Rankings to label
map_path = '../data/beerslaw/experiment_keys/permutation_maps/vector_binary.yaml'
with open(map_path) as fp:
    vectorranking_map = yaml.load(fp, Loader=yaml.FullLoader)
    
map_path = '../data/beerslaw/experiment_keys/permutation_maps/nconcepts_binary.yaml'
with open(map_path) as fp:
    binaryranking_map = yaml.load(fp, Loader=yaml.FullLoader)
```

```python
rankings['concepts_binary'] = rankings['ranking'].apply(lambda x: vectorranking_map['map'][x])
rankings['binconcepts'] = rankings['ranking'].apply(lambda x: binaryranking_map['map'][x])
```

```python
with open('../data/beerslaw/experiment_keys/over10.pkl', 'rb') as fp:
    over10 = pickle.load(fp)
    
edmrankings = rankings.reset_index()
edmrankings = edmrankings[edmrankings['username'].isin(over10)]
```

# General Demographics

```python
n = len(rankings)
demographics = ['language', 'gender', 'field', 'year']
fig, axes = plt.subplots(4, 1, figsize=(6, 10))

languages = rankings['language'].value_counts()
axes[0].bar([0, 1], [languages['Français']/n, languages['Deutsch']/n], color='#0C1618')
axes[0].set_xticks([0, 1])
axes[0].set_xticklabels(['French', 'German'])

field = rankings['field'].value_counts()
axes[1].bar([0, 1, 2, 3, 4], [
    field['Chemistry']/n, field['Chemistry, Textiles']/n, field['Biology']/n, 
    field['Pharma Chemistry']/n, field['Fast track']/n
], color='#3F7267')
axes[1].set_xticks([0, 1, 2, 3, 4])
axes[1].set_xticklabels(['chemistry', 'textiles', 'biology', 'pharma', 'fast'])

gender = rankings['gender'].value_counts()
axes[2].bar([0, 1, 2, 3], [
    gender[1]/n, gender[2]/n,
    gender[3]/n, gender[4]/n
], color='#E6D06A')
axes[2].set_xticks([0, 1, 2, 3])
axes[2].set_xticklabels(['male', 'female', 'other', 'NA'])

year = rankings['year'].value_counts()
axes[3].bar([0, 1, 2], [
    year['1st']/n, year['2nd']/n,
    year['3rd']/n
], color='#D1AC00')
axes[3].set_xticks([0, 1, 2])
axes[3].set_xticklabels(['1st', '2nd', '3rd'])

for i in range(4):
    axes[i].set_ylim([0, 1])
```
