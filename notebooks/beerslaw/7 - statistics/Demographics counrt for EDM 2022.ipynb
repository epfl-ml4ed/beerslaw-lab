```python
map_path = '../data/experiment_keys/permutation_maps/vector_binary.yaml'
with open(map_path) as fp:
    mapping = yaml.load(fp, Loader=yaml.FullLoader)
```

```python
genders = []
years = []
fields = []
languages = []
permutations = []
last_ts = []
for idx in config['id_dictionary']['sequences']:
    with open(config['id_dictionary']['sequences'][idx]['path'], 'rb') as fp:
        sim = pickle.load(fp)
    lid = config['id_dictionary']['sequences'][idx]['learner_id']
    genders.append(post_test.loc[lid]['gender'])
    years.append(post_test.loc[lid]['year'])
    fields.append(post_test.loc[lid]['field'])
    languages.append(post_test.loc[lid]['language'])
    permutations.append(mapping['map'][post_test.loc[lid]['ranking']])
    last_ts.append(sim['last_timestamp'])
    
```

```python
np.mean(last_ts)
```

```python
data = {
    'gender': genders,
    'year': years,
    'field': fields,
    'language': languages,
    'permutation': permutations
}
demographics = ['gender', 'year', 'field', 'language', 'permutation']
```

```python
def count_concepts(permutation:str):
    intermediate = int(permutation[0]) + int(permutation[1]) + int(permutation[2]) 
    return int(intermediate >= 2)

for demo in demographics: 
    print('Demographics: {}'.format(demo))
    for attribute in np.unique(data[demo]):
        indexes = [idx for idx in range(len(data[demo])) if data[demo][idx] == attribute]
        permus = [permutations[idx] for idx in indexes]
        permus = [count_concepts(permu) for permu in permus]
        count = Counter(permus)
        print('   attribute {}: {}'.format(attribute, count))
    
```

```python
test_indexes = []
for i in range(10):
    test_indexes = test_indexes + list(res[i]['test_index'])
test_indices = [res['indices'][idx] for idx in test_indexes]
test_ys = [res['y'][idx] for idx in test_indexes]
```

```python
lids = [cap_config['id_dictionary']['sequences'][idx] for idx in test_indices]
```

```python
path = '/Users/cock/kDrive/PhD/Projects/Labs/beerslaw-lab/data/post_test/Phys_CapacitorLab_cleaned.xlsx'
sim_details = pd.read_excel(path)
```

```python
matching_df = sim_details[['Metacog ID', 'Class']].dropna()
matching_dic = {}
for i, row in matching_df.iterrows():
    lid = str(int(row['Metacog ID']))
    matching_dic[lid] = row['Class']
```

```python
classes = [matching_dic[lid['learner_id']] for lid in lids]
# test_ys
for curclass in np.unique(classes):
    indices = [i for i in range(len(classes)) if classes[i] == curclass]
    ys = [test_ys[idx] for idx in indices]
    print('Number of high learners in {}: {} [{}]'.format(curclass, Counter(ys), len(indices)))
```
