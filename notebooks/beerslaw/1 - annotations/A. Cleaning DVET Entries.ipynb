This notebook is used to clean entries that were made by the D-VET Team

```python
import os
import numpy as np
import pandas as pd
pd.set_option('display.max_columns', None)
```

# Data

```python
annotated = pd.read_csv('../../data/post_test/annotatedv0.csv', sep='\t', index_col=0)
annotated['label_seven'].unique()
```

```python
extracted = pd.read_csv('../../data/post_test/extracted.csv', sep='\t', index_col=0)
```

# Cleaning

```python
# Checking those suspected to be D-VET
dvet = annotated[annotated['label_seven'] == 'DVET']
print(list(dvet['username'].unique()))
dvet = extracted[extracted['username'].isin(dvet['username'].unique())]
dvet

## Taking out all of the entries except the one with a sensible feedback
```

```python
# Checking those suspected to be D-VET
dvet = annotated[annotated['label_seven'] == 'to check']
print(list(dvet['username'].unique()))
dvet = extracted[extracted['username'].isin(dvet['username'].unique())]
dvet

## Entries looked "sound". The weirdest one is probably from a student who cheated on their classmates 
## but did not copy their answers over correctly
```

```python
# Checking those suspected to be unrelated because misaligned
dvet = annotated[annotated['label_seven'] == 'unrelated']
print(list(dvet['username'].unique()))
dvet = extracted[extracted['username'].isin(dvet['username'].unique())]
dvet

## Weird entries appear to be from confused people
```
