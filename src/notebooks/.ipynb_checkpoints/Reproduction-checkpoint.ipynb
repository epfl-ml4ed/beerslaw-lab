### Fold reconstruction

```python
with open('../../data/reproduction/2classes_sgenc_partialfolds.pkl', 'rb') as fp:
    old = pickle.load(fp)
```

```python
files = os.listdir(root)
files = [f for f in files if 'ann_fembeddingscorrect-closed_ts' in f]
files = [f for f in files if 'predictions' not in f and '500' not in f]

new_ts = {}
for f in files:
    regex = re.compile('ts([0-9]*)')
    ts = int(regex.findall(f)[0])
    with open(root + f, 'rb') as fp:
        results = pickle.load(fp)
        lengths = [len(x) for x in results['x']]
        new_ts[ts] = {
            'w1' : results['w1'],
            'map_events' : results['map_events'],
            'x' : results['x'],
            'y' : results['y'],
            'ids' : results['ids'],
            'map_labels' : results['map_labels'],
            'raw_x' : results['raw_x'],
            'folds': {}
        }
        
        for fold in results['folds']:
            new_ts[ts]['folds'][fold] = {
                'train_index' : results['folds'][fold]['train_index'],
                'x_train': results['folds'][fold]['X_train'],
                'y_train': results['folds'][fold]['y_train'],
                'test_index': results['folds'][fold]['test_index'],
                'x_test': results['folds'][fold]['X_test'],
                'y_test': results['folds'][fold]['y_test'],
                'x_oversampled': results['folds'][fold]['oversampled_X_train'],
                'y_oversampled': results['folds'][fold]['oversampled_y_train'],
                'x_long_test': results['folds'][fold]['longenough_X_test'],
                'y_long_test': results['folds'][fold]['longenough_y_test'],
            }
            lens = [lengths[i] for i in results['folds'][fold]['test_index']]
            long_indices = [i for i in range(len(lens)) if lens[i] > ts]
            long_indices = [results['folds'][fold]['test_index'][i] for i in long_indices]
            new_ts[ts]['folds'][fold]['long_indices'] = long_indices
            
with open('../../data/reproduction/2classes_sgenc_partialfolds.pkl', 'wb') as fp:
    pickle.dump(new_ts, fp)
```

```python
bug = 0
non_bug = 0

for f in files:
    with open('../../data/reproduction/' + f, 'rb') as fp:
        partial = pickle.load(fp)
    for l in partial:
        for fold in partial[l]['folds']:
            basis = partial[l]['folds'][fold]
            if len(basis['x_long_test']) != len(basis['y_long_test']) or len(basis['y_long_test']) != len(basis['long_indices']):
#                 continue
                print('file {} with length {} on fold {} has a problem'.format(f, l, fold))
                print('   ', len(basis['x_long_test']), len(basis['y_long_test']), len(basis['long_indices']))
                bug += 1
                
            else:
                non_bug += 1
#                 print('file {} with length {} on fold {} has a problem'.format(f, l, fold))
```
