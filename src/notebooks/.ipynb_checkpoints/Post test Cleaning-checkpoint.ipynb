In this notebook, we take the extracted post test information, and we clean it such that: \
    - DVET entries are cleaned out \
    - missing rankings are taken away

```python
import os 
import re

import numpy as np
import pandas as pd
pd.set_option('display.max_columns', None)
```

# Data

```python
post = pd.read_csv('../../data/post_test/extracted.csv', sep='\t', index_col=0)
```

# Clean

```python
print('length of the dataframe: {}, unique usernames: {}'.format(len(post), len(post['username'].unique())))
```

```python
# list of usernames that should be excluded
status_rejected = [
    "'name'", "'1'"
]

rejected = post[post['status'].isin(status_rejected)]
usernames_rejected = list(rejected['username'])
```

```python
# remove non started entries
post = post[post['status'] != "'name'"]

# remove test entries (only username)
post = post[post['status'] != "'1'"]
post = post[]
```
