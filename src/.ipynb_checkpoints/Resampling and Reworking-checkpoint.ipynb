# Sequence building
In temp, we only keep the answers from the ranking, and the associated metacog id. <br>
We build the sequence such that:
- the entry in column cq1_1 is the configuration they ranked first
- the entry in column cq1_2 is the configuration they ranked second
- the entry in column cq1_3 is the configuration they ranked third
- the entry in column cq1_4 is the configuration they ranked fourth


```python
temp = phys[['metacog_id', 'cq1_1', 'cq1_2', 'cq1_3', 'cq1_4']]
temp = temp.dropna()
temp = temp.replace('1 (greatest amount of stored energy)', '1')
temp = temp.replace('4 (least amount of stored energy)', '4')

def getSequence(row):
    rank = {}
    rank[row['cq1_1']] = '1'
    rank[row['cq1_2']] = '2'
    rank[row['cq1_3']] = '3'
    rank[row['cq1_4']] = '4'
    string = ''
    if '1' in rank:
        string = string + rank['1']
    if '2' in rank:
        string = string + rank['2']
    if '3' in rank:
        string = string + rank['3']
    if '4' in rank:
        string = string + rank['4']
    return string

temp['sequence'] = temp.apply(getSequence, axis=1)
```

# Counts, per permutations, of users in the logs AND the post test

**possible_ids** is a dictionary whose keys are the ids of the users in the logs, and whose values are the logs corresponding to the "key" user. 

```python
print(len(possible_ids))
```

```python
possible_ids
```

**id_sequence** is a dictionary whose keys are the ids in the post test, and whose value is the ranking answer of the "key" student

```python
id_sequence = {}
for i in range(len(temp)):
    id_sequence[temp.iloc[i]['metacog_id']] = temp.iloc[i]['sequence']
print(id_sequence)
```

**matchings** is a dictionary containing all the user ids as their key. The values are the ranking given, and the amount of files that user can be found in in the logs.

```python
matchings = {}
for iid in id_sequence:
    matchings[iid] = {}
    matchings[iid]['sequence'] = id_sequence[iid]
    if iid in possible_ids:
        matchings[iid]['files'] = possible_ids[iid]
    else:
        matchings[iid]['files'] = []
```

**in_counter** is a dictionary which counts the amount of ids that was in both the logs and the post test, per permutation <br>

**not counter** is a dictionary which counts the amount of ids that was not in the logs but only in the post test, per permutation

```python
in_counter = Counter()
in_ids = []
not_counter = Counter()
not_ids = []
multiples = {}
singles = {}
for iid in matchings:
    if len(matchings[iid]['files']) == 0:
        not_counter[matchings[iid]['sequence']] += 1
    elif len(matchings[iid]['files']) == 1:
        in_counter[matchings[iid]['sequence']] += 1
        singles[iid] = {}
        singles[iid]['files'] = matchings[iid]['files']
        singles[iid]['sequence'] = matchings[iid]['sequence']
    else:
        multiples[iid] = {}
        multiples[iid]['files'] = matchings[iid]['files']
        multiples[iid]['sequence'] = matchings[iid]['sequence']
```

# Resampling

```python
fs = FakeSampler.FakeSampler()
states = ['other', 'se_other', 'se_bc_other', 'bc_other', 'voltage', 'se_voltage', 'se_bc_voltage', 'bc_voltage', 'plateseparation', 'se_plateseparation', 'se_bc_plateseparation', 'bc_plateseparation', 'platearea', 'se_platearea', 'se_bc_platearea', 'bc_platearea', 'break', 'se_break', 'se_bc_break', 'bc_break']
sampler = AdaptiveSampler.AdaptiveSampler()
```

```python
in_dir = '../Objects/simulations/news/'
files = os.listdir(in_dir)

out_dir = '../Objects/timelines/adaptive/'
keys = ['cumulative60', 'cumulative80']
max80 = 0
max60 = 0

id_timelines = {}
for file in files:
    path = in_dir + file
    with open(path, 'rb') as fp:
        sim = pickle.load(fp)
        br = fs.getBreaks(sim, [], states, 0.6)
    timeline = AdaptiveTimeline.AdaptiveTimeline(sim.user.permutation, sim.user.learner_id, out_dir, '')
    timeline.simpath = path
    
    cumulative_threshold = 0.8
    seq, breaks = sampler.adapt_sample(sim, br, states, cumulative_threshold)
    timeline.add_timeline('cumulative60', seq, {})
    timeline.add_breaks('cumulative60', breaks)
    if len(seq) > max80:
        max80 = len(seq)
    
    cumulative_threshold = 0.6
    seq, breaks = sampler.adapt_sample(sim, br, states, cumulative_threshold)
    timeline.add_timeline('cumulative80', seq, {})
    timeline.add_breaks('cumulative80', breaks)
    if len(seq) > max60:
        max60 = len(seq)
    
    timeline.save()
    id_timelines[str(sim.user.learner_id)] = timeline

with open('../Objects/data keys/id_timelinescumulative60.pkl', 'wb') as fp:
    pickle.dump(id_timelines, fp)

with open('../Objects/data keys/id_timelinescumulative80.pkl', 'wb') as fp:
    pickle.dump(id_timelines, fp)
```
