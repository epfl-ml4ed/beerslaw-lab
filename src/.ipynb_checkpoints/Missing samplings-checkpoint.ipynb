```python
news = {}
raw_perms = []
for idd in maps:
    key = str(idd)
    key = key[:key.find('.')]
    news[key] = {}
    news[key]['sequence'] = maps[idd]
    news[key]['files'] = []
    raw_perms.append(idd)
```

# User ids

```python
possible_ids = {}
```

```python
directories = ['../Data/Phys_1120_Comparison/', '../Data/Phys_1120_Explore/', '../Data/Phys_1125_Comparison/', '../Data/Phys_1125_Explore/']
for directory in directories:
        for filename in os.listdir(directory):
            if filename.endswith('.txt'):
                print(filename)
                path = directory + filename
                with open(path, 'r') as fp:
                    user_log = json.load(fp)
                    lid = str(int(user_log['session']['learner_id'].replace('NaN', '')))
                    if lid in possible_ids:
                        possible_ids[lid].append(path)
                    else:
                        possible_ids[lid] = [path]
```

```python
from collections import Counter
```

```python
counts = Counter()
misses = Counter()
multiples = {}
singles = {}
missings = []
for iid in news:
    if len(news[iid]['files']) != 0:
        counts[''.join(news[iid]['sequence'])] += 1
        singles[iid] = {}
        singles[iid]['files'] = news[iid]['files']
        singles[iid]['sequence'] = ''.join(news[iid]['sequence'])
    if len(news[iid]['files']) > 1:
        multiples[iid] = {}
        multiples[iid]['files'] = news[iid]['files']
        multiples[iid]['sequence'] = ''.join(news[iid]['sequence'])
    else:
        missings.append(iid)
        misses[''.join(news[iid]['sequence'])] += 1
        
```
