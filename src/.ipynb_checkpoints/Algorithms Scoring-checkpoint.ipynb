```python
path = '../Objects/summaries/DL/'
```

```python
def readSummary(summary):
    with open(path + summary, 'rb') as fp:
        summary = pickle.load(fp)
    summary.viewScores('all_test_average_accuracy_mean')
    summary.save()
    return summary
```

```python
def retrieveColumns(keywords, df, index):
    columns = list(df.columns)
    temp = columns[:index]
    for k in keywords:
        columns = [x for x in columns if k in x]
    columns = temp + columns
    return df[columns]
            
def getAllAverageScores(summary, parameters):
    columns = parameters + [
            'all_test_average_accuracy_mean',
            'all_test_average_precision_mean', 
            'all_test_average_recall_mean', 
            'all_test_average_mcc_mean',
        ]
    temp = summary.scores[columns]
    temp.columns = parameters + ['accuracy', 'precision', 'recall', 'mcc']
    return temp
```

```python
def selectBestScore(df, index, n):
    df = pd.DataFrame(df.iloc[index]).sort_values(318, ascending=False)
    for i in range(n):
        print(df.iloc[i])
```

```python
def retrieveColumns(keywords, df, index):
    columns = list(df.columns)
    temp = columns[:index]
    for k in keywords:
        columns = [x for x in columns if k in x]
    columns = temp + columns
    return df[columns]
            
def getAllAverageScores(summary, parameters):
    columns = parameters + [
            'all_test_average_accuracy_mean',
            'all_test_average_precision_mean', 
            'all_test_average_recall_mean', 
        ] + [
            'all_test_inverse_weighted_accuracy_mean',
            'all_test_inverse_weighted_precision_mean', 
            'all_test_inverse_weighted_recall_mean', 
        ]
    temp = summary.scores[columns]
    temp.columns = parameters + [
        'accuracy (av)', 'precision (av)', 'recall (av)', 'accuracy (iw)', 'precision (iw)', 'recall (iw)']
    return temp
```

```python
def retrieveColumns(keywords, df, index):
    columns = list(df.columns)
    temp = columns[:index]
    for k in keywords:
        columns = [x for x in columns if k in x]
    columns = temp + columns
    return df[columns]
            
def getAllAverageScores(summary, parameters):
    columns = parameters + [
            'all_test_average_accuracy_mean',
            'all_test_average_recall_mean', 
            'all_test_average_mcc_mean', 
        ] + [
            'all_test_inverse_weighted_accuracy_mean',
            'all_test_inverse_weighted_recall_mean',
            'all_test_inverse_weighted_mcc_mean',
        ]
    temp = summary.scores[columns]
    temp.columns = parameters + [
        'accuracy (av)', 'recall (av)', 'mcc (av)',
        'accuracy (iw)', 'recall (iw)', 'mcc (iw)']
    return temp
```

```python
in_directory = '../Objects/timelines/adaptive/'
files = os.listdir(in_directory)
```

```python
length_sequences = Counter()
for file in files:
    path = in_directory + file
    with open(path, 'rb') as fp:
        timeline = pickle.load(fp)
        length_sequences[len(timeline.timelines['cumulative80'])] += 1
```

```python
length_sequences
```

```python
milestones = [
    10, 20, 30, 40, 50, 60, 80, 90, 100, 110, 120, 140, 160, 180, 200, 220, 240
]
```

```python
ms = Counter()
for m in milestones:
    for l in length_sequences:
        if l >= m:
            ms[m] += length_sequences[l]
    
```

```python
total = 0
for l in length_sequences:
    total += length_sequences[l]
```

```python
for m in ms:
    ms[m] = ms[m] / total
```
