```python
def printDecisions(algorithm, name):
    i = 0
    for estimator in algorithm.model.estimators_:
        plt.figure(figsize=(20, 30))
        tree.plot_tree(estimator, feature_names=algorithm.state_list, 
                       class_names=list(algorithm.target_index.keys()),
                      fontsize=15)
        path = '../Figures/results/RandomForestTrees/' + name + '_tree' + str(i) + 'png'
        i += 1
        plt.savefig(path)
        plt.show()
```

```python
state_settings = {
            'feature_type' : 'static',
            'state_list': ['other', 'se_other', 'se_bc_other', 'bc_other', 'voltage', 'se_voltage', 'se_bc_voltage', 'bc_voltage', 'plateseparation', 'se_plateseparation', 'se_bc_plateseparation', 'bc_plateseparation', 'platearea', 'se_platearea', 'se_bc_platearea', 'bc_platearea', 'break', 'se_break', 'se_bc_break', 'bc_break'],
            'permutation_list' : ['1324', '4213', '1234', '4231', '2413'],
    }
performance_measures = ['accuracy', 'recall', 'mcc', 'precision']
averageing = ['average', 'weighted', 'inverse_weighted']

############################################# Parameters for flatten and cumulative
algo_settings = {
    'key' : 'count',
    'n_estimators' : 7,
    'criterion' : 'gini',
    'max_depth': 8,
    'min_samples_split': 5,
    'max_features': 'auto',
    'nfolds':7,
    'training_size':0.8,
    'test_size':0.2,
    'len_sequences' : 240,
    'length_type' : 'padding'
}

algorithm = PermutationRF.PermutationRF('count_final', algo_settings, state_settings)
for i in range(1):
    train, test = algorithm.get_cv_data(i)
    algorithm.train(train)
    print('* * ' * 30)
    algorithm.predict(test[0])
    print()
    algorithm.predict(train[0])
    
    
    printDecisions(algorithm, 'count_test_permutationclassifier')

algorithm.save()
```
