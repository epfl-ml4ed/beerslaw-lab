```python
def checkNan(x):
    if type(x) == str:
        return False
    elif np.isnan(x):
        return True
    else:
        return 'bruh'
    
def getGroup(row):
    if checkNan(row['eq_1']) and not checkNan(row['cq_1_1']):
        return 'ranking'
    elif checkNan(row['cq_1_1']) and not checkNan(row['eq_1']):
        return 'control'
    else:
        return 'not_completed'
```

```python
ans1120['group'] = ans1120.apply(lambda x: getGroup(x), axis=1)
ans1120 = pd.get_dummies(ans1120, columns=['group'])
```

```python
control = ans1120[ans1120['group_control'] == 1]
```

```python
def process_list(y):
    temp = [porter_stemmer.stem(wordnet_lemmatizer.lemmatize(x)) for x in y]
    return ' '.join(temp)

def word_in_text(words, text):
    words = re.sub('s+','s*', '|'.join(words)) # Find pattern, replace with other, the list
    text = text.lower()
    match = re.search(words, text)
    if match:
        return True
    return False
```

## Question 1 - control : When the battery is CONNECTED, explain how changing each factor changes the amount of stored energy. Use data from the simulation to support your claim.

```python
control['eq1_raw'] = control['eq_1']
control['eq_1'] = control['eq_1'].str.lower()

wordcloud = WordCloud(background_color="white").generate(' '.join(control['eq_1']))

plt.figure(figsize=(10, 5))
plt.imshow(wordcloud, interpolation='bilinear')
plt.axis("off")
plt.show()
```

```python
wordnet_lemmatizer = WordNetLemmatizer()
porter_stemmer  = SnowballStemmer('english')

factor_analysed = [
    'storage', 'store', 'plate', 'area', 'separation', 'voltage', 'battery'
]
factor_analysed = [porter_stemmer.stem(wordnet_lemmatizer.lemmatize(x)) for x in factor_analysed]

actions = [
    'changing'
]
actions = [porter_stemmer.stem(wordnet_lemmatizer.lemmatize(x)) for x in actions]
relativity = [
    'proportionally', 'increase', 'decrease', 'higher', 'lower', 'closer', 'less', 'more', 'amount', 'size', 'affect',
    'larger', r'\+', '-', 'positive', 'negative', 'reducing', 'greatest', 'smallest', 'sign'
]
relativity = [porter_stemmer.stem(wordnet_lemmatizer.lemmatize(x)) for x in relativity]

quantity = [
    r'[0-9]*', 'mm', '^2', 'pj', 'v', 'maximum', 'minimum', r'[0-9]/[0-9]', '([0-9]|\s)*x', r'[0-9]+pj', 'pico', 'constant'
]
quantity = [porter_stemmer.stem(wordnet_lemmatizer.lemmatize(x)) for x in quantity]

buzzes = [factor_analysed, actions, relativity, quantity]
buzz_columns = ['factor_identifications', 'actions', 'relations', 'quantitisation']
```

```python
temp['eq1_processed'] = temp['eq_1'].apply(lambda x: word_tokenize(x))
temp['eq1_processed'] = temp['eq1_processed'].apply(lambda x: process_list(x))
for liste in buzzes:
    for w in liste:
        temp[w] = temp['eq1_processed'].apply(lambda answer: word_in_text([w], answer)).astype(int)
```

```python
def getScore(row, col):
    temp = row[col]
    return sum(temp) / len(col)
```

```python
for i in range(len(buzzes)):
    buzz = buzzes[i]
    buzz_name = buzz_columns[i]
    
    temp[buzz_name] = temp.apply(lambda x : getScore(x, buzz), axis=1)
```
