# States 17

```python
path = '../data/sequenced simulations/17s/'
files = os.listdir(path)
files = [f for f in files if '_sequenced' in f]
```

```python
break_filter = Cumul60BreakFilter()
```

```python
simseqs = {}
for index, file in enumerate(files):
    with open(path + file, 'rb') as fp:
        simseq = pickle.load(fp)
        
    seq, begin, end = simseq['sequence'], simseq['begin'], simseq['end']
    seq, begin, end = break_filter.inpute_all_breaks(seq, begin, end)
    simseqs[index] = {
                    'sequence': seq,
                    'begin': begin,
                    'end': end,
                    'permutation' : simseq['permutation']
                    }
```

## Entire dataset

### Lengths of sequences

#### Timesteps

```python
lengths = []
for index in simseqs:
    lengths.append(len(simseqs[index]['sequence']))
sns.displot(lengths)
plt.title('Distribution of the time steps lengths')
plt.show()
```

#### Duration

```python
def s_to_min(x, intervals):
    maxi = max(x)
    seconds = list(range(0, int(maxi+60), intervals))
    minutes = list(np.array(range(0, len(seconds), 1)) * (intervals / 60))
    return seconds, minutes
ends = []
for index in simseqs:
    ends.append((simseqs[index]['end'][-1]))
sns.displot(ends)
s, m = s_to_min(ends, 3600)
plt.xticks(s, m)
plt.xlabel('minutes')
plt.title('Distribution of the duration lengths')
plt.show()

smaller_ends = []
for index in simseqs:
    smaller_ends.append((simseqs[index]['end'][-1]))
smaller_ends = [se for se in ends if se < 3600]
sns.displot(smaller_ends)
s, m = s_to_min(smaller_ends, 600)
plt.xticks(s, m)
plt.xlabel('minutes')
plt.title('Distribution of the duration lengths')
plt.show()

pd.Series(ends).describe()
```

### Distribution of events

```python
events = []
for index in simseqs:
    events = events + simseqs[index]['sequence']
    
sns.displot(events)
plt.title('Distribution of events across the sequences')
plt.xticks(rotation=90)
plt.show()
```

## 2 Classes

```python
label_map = '../data/experiment keys/permutation_maps/2class_map.yaml'
with open(label_map, 'r') as f:
    label_map = yaml.load(f, Loader=yaml.FullLoader)
    
label_dist = []
label_seqs = {}
for index in simseqs:
    perm = simseqs[index]['permutation']
    if perm in label_map['map']:
        lab = label_map['map'][perm]
        label_dist.append(lab)
        if lab in label_seqs:
            label_seqs[lab].append(simseqs[index])
        else:
            label_seqs[lab] = [simseqs[index]]
    
```
